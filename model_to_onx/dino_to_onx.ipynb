```python
import torch
import groundingdino.models.GroundingDINO as GroundingDINO

# Load the model
model = GroundingDINO.build_model()
checkpoint = torch.load("grounding-dino-base.pth", map_location="cpu")
model.load_state_dict(checkpoint['model'])
model.eval()

# Dummy input for ONNX export
dummy_input = torch.randn(1, 3, 640, 640)

# Export to ONNX
torch.onnx.export(
    model, dummy_input, "grounding-dino.onnx",
    opset_version=11,
    input_names=["input"],
    output_names=["output"]
)
print("Model exported to ONNX format.")
```
